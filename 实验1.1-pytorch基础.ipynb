```python
## pytorch的基础操作
### 构建矩阵
####构建全0的tensor矩阵
x1= torch.zeros(5,3,dtype=torch.long)
print(x1)
print("----"*20)
x1
```

```python
#### 构建随机的tensor矩阵
x2= torch.rand(3,4)
print(x2)
print("----"*20)

#### 构建全1的tensor矩阵
x3= torch.ones(3,4).int()# 指定类型
print(x3)
print("----"*20)

### 从数据直接构建tensor
x4= torch.tensor([2,2,2]).int()
print(x4)
```

```python
### 从已有的tensor构建一个tensor。这些方法会重用原来tensor的特征，例如，数据类型，除非提供新的数据。
x5= x4.new_ones(5,3)#注意不能从一维tensor直接构建，tensor没有torch属性
print(x5)
print("----"*20)

### 根据已有tensor的维度随机生成一个tensor
x6= torch.randn_like(x5,dtype=torch.float)
print(x6)
```

```python
## tensor的形状
x6.size()#注意torch.Size`返回的是一个tuple
```

```python
## tensor的运算：加减乘除
### 加法1
y = torch.ones(5,3)
x = torch.ones(5,3)
z=x+y
z
```

```python
### 加法2
torch.add(x, y)
```

```python
### 加法3  in-place加法  在原有的基础上加,这里改变了y
print(y.add_(x))
print("----"*20)

z=x+y 
print(z)
```

```python
## tensor索引操作同numpy
z[1:,1:]
```

```python
## Resizing: 如果你希望resize/reshape一个tensor，可以使用torch.view改变维度
print(z.view(3,-1))# -1代表该维度可以计算得出
z.view(-1,15)
```

```python
## 其他重要操作
### 一个数值改成python数值
x = torch.randn(1)
print(x)
print("----"*20)
print(x.item())
```

```python
### 转置
u=torch.ones(5,2)
print(u)
print("----"*20)
print(u.transpose(0,1))
```
